```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
 


def run_selenium(start_Year, end_Year):
    Population_start_Year = {'Country': [], 'Population': []}
    df_Population_start_Year = pd.DataFrame(data=Population_start_Year)
    
    Population_end_Year = {'Country': [], 'Population': []}
    df_Population_end_Year = pd.DataFrame(data=Population_end_Year)
    
    
    browser = webdriver.Safari(executable_path = '/usr/bin/safaridriver')  
    browser.get('https://data.worldbank.org/indicator/sp.pop.totl?end=' + str(end_Year) + '&start=' + str(start_Year))
    time.sleep(5)
    
    
    #cty = browser.find_element_by_xpath('//*[@id="main"]/div/div/article/section/div/div/div/section/div/div[2]/div[1]/div[1]')
    #print(cty.text)
            #//*[@id="main"]/div/div/article/section/div/div/div/section/div/div[2]/div[1]/div[1]
    Xpath = '//*[@id="main"]/div/div/article/section/div/div/div/section/div/div[2]/div[' 
    cty_set = set()
    for i in range(1, 249):
        
        XPATH_Country = Xpath + str(i) + ']/div[1]'
        
        
        cty = browser.find_element_by_xpath(XPATH_Country).text
        if cty in Countries.keys():
            #print(cty)
            
            XPATH_Year_start = Xpath + str(i) + "]/div[2]"
            XPATH_Year_end = Xpath + str(i) + "]/div[3]"
            
            pop_Year_start = browser.find_element_by_xpath(XPATH_Year_start).text
            pop_Year_end = browser.find_element_by_xpath(XPATH_Year_end).text
            cty_shortName = Countries[cty]
            cty_set.add(cty_shortName)
            
            pop_Year_start = locale.atof(pop_Year_start)
            pop_Year_end = locale.atof(pop_Year_end)
 
            cty_shortName = Countries[cty]
            print(cty_shortName, pop_Year_start, pop_Year_end )
           
            
            df_Population_start_Year = df_Population_start_Year.append({'Country': cty_shortName,'Population':float(pop_Year_start) }, ignore_index=True)
            df_Population_end_Year = df_Population_end_Year.append({'Country': cty_shortName,'Population':float(pop_Year_end) }, ignore_index=True)
    
    #print(df_Population_start_Year)
    #print(df_Population_end_Year)
    
    df_Population_start_Year.to_csv (path_or_buf= '/Users/mac/Downloads/population_' + str(start_Year) + '.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
    df_Population_end_Year.to_csv (path_or_buf= '/Users/mac/Downloads/population_' + str(end_Year) + '.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path

    print(country_set.difference(cty_set))
    browser.quit()
```

```python
for year in range(2000, 2015, 2):
    run_selenium(year,year+1)
```
