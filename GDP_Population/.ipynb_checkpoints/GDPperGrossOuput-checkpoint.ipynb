```python
import pandas as pd
```

```python
GDPperGrossOuput = {'Country': []}
```

```python
GDPperGrossOuput =  pd.DataFrame(GDPperGrossOuput)
```

```python
for year in range(2000, 2015, 1):

    #df = pd.read_csv('population_' + str(year) + '.csv')
    df = pd.read_csv('GdpPop' + str(year) + '.csv')

    country_name = list(x for x in df['Country'])

    GDPperGrossOuput['Country'] = country_name

    GDPperOutput = list(x for x in df['GDPperGrossOut'])


    GDPperGrossOuput[str(year)] = GDPperOutput

    
```

```python

GDPperGrossOuput
```

```python
GDPperGrossOuput.to_csv (path_or_buf= '../project_data/gdp_per_gross_output.csv',  index= None, header=True)
```
